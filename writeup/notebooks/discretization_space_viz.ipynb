```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta = 0.01
x = np.arange(-1.0, 1.0, delta)
y = np.arange(-1.0, 1.0, delta)
X, Y = np.meshgrid(x, y)
Z = 0.5*((X - 0.3)**2 + (Y-0.1)**2)

fig, ax = plt.subplots()
ax.set_aspect(1.0)
ax.scatter([0.3], [0.1], marker='x', color='tab:blue')
cs = ax.contour(X, Y, Z, linestyles='dashed', colors='tab:gray', levels=10)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_xticks([])
ax.set_yticks([])

# Plot the solution-space
# I'm just making up a curve here, which should have two closest-local-approaches to the data-minimum.
theta = 2*np.pi*(35/360)
a = np.sin(theta)
b = np.cos(theta)
x0 = -0.15
y0 = -0.7
c = 5
z2_grid = np.arange(-1, +1, 0.01)
x_grid = a*c*z2_grid**2 + b*z2_grid + x0
y_grid = (b*c*z2_grid**2 - a*z2_grid)/(b**2 - a**2) + y0
ax.plot(x_grid, y_grid, ls='-', lw=2.0, color='tab:green')

z2_min1 = 0.33
ax.scatter(a*c*z2_min1**2 + b*z2_min1 + x0, (b*c*z2_min1**2 - a*z2_min1)/(b**2 - a**2) + y0, zorder=2, color='tab:orange')
z2_min2 = -0.205
ax.scatter(a*c*z2_min2**2 + b*z2_min2 + x0, (b*c*z2_min2**2 - a*z2_min2)/(b**2 - a**2) + y0, zorder=2, edgecolors='tab:orange', facecolors='none')

plt.show()

fig.savefig('disc_space_viz.png')
```
